```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../DATA/dog_backpack.png')
```

```python
plt.imshow(img)
```

```python
img_rbg = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rbg)
```

```python
new_img = img_rbg.copy()
```

```python
new_img = cv2.flip(new_img,0)
plt.imshow(new_img)
```

```python
cv2.rectangle(img_rbg,pt1=(200,380),pt2=(600,700),color=(255,0,0),thickness=10)
plt.imshow(img_rbg)
```

```python
vertices = np.array([[250,700],[425,400],[600,700]],np.int32)
vertices
```

```python
vertices.shape
```

```python
pts = vertices.reshape((-1,1,2))
pts.shape
```

```python
cv2.polylines(img_rbg,[pts],isClosed=True,color=(0,0,255),thickness=10)
plt.imshow(img_rbg)
```

```python
def create_circle(event,x,y,flags,param):
    if event == cv2.EVENT_RBUTTONDOWN:
        cv2.circle(img,(x,y),100,(0,0,255),thickness=10)
        
img = cv2.imread("../DATA/dog_backpack.png")
cv2.namedWindow(winname='dog')
cv2.setMouseCallback('dog',create_circle)

while True:
    cv2.imshow('dog',img)
    if cv2.waitKey(20) & 0xFF == 27:
         break
            
cv2.destroyAllWindows()
```
