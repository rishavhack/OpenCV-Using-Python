```python
gorilla = cv2.imread('../DATA/gorilla.jpg',0)
```

```python
def display_img(img,cmap=None):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
display_img(gorilla,cmap='gray')
```

```python
gorilla.shape
```

```python
hist_vlaues = cv2.calcHist([gorilla],channels=[0],mask=None,histSize = [256],ranges=[0,256])
```

```python
plt.plot(hist_vlaues)
```

```python
eq_gorilla = cv2.equalizeHist(gorilla)
display_img(eq_gorilla,cmap='gray')
```

```python
hist_vlaues = cv2.calcHist([eq_gorilla],channels=[0],mask=None,histSize = [256],ranges=[0,256])
```

```python
plt.plot(hist_vlaues)
```
