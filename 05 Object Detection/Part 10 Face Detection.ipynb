```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
nadia = cv2.imread('../DATA/Nadia_Murad.jpg',0)
denis = cv2.imread('../DATA/Denis_Mukwege.jpg',0)
solvay = cv2.imread('../DATA/solvay_conference.jpg',0)
```

```python
#plt.imshow(solvay,cmap='gray')
```

```python
face_cascade = cv2.CascadeClassifier('../DATA/haarcascades/haarcascade_frontalface_default.xml')
```

```python
def detect_img(img):
    
    face_img = img.copy()
    
    face_rects = face_cascade.detectMultiScale(face_img)
    
    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
        
    return face_img
```

```python
result = detect_img(denis)
```

```python
plt.imshow(result,cmap='gray')
```

```python
result1 = detect_img(nadia)
plt.imshow(result1,cmap='gray')
```

```python
result2 = detect_img(solvay)
plt.imshow(result2,cmap='gray')
```

```python
def adj_detect_img(img):
    
    face_img = img.copy()
    
    face_rects = face_cascade.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5)
    
    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
        
    return face_img
```

```python
result3 = adj_detect_img(solvay)
plt.imshow(result3,cmap='gray')
```

```python
eye_cascade = cv2.CascadeClassifier('../DATA/haarcascades/haarcascade_eye.xml')
```

```python
def detect_eye(img):
    
    face_img = img.copy()
    
    eye_rects = eye_cascade.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5)
    
    for (x,y,w,h) in eye_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
        
    return face_img
```

```python
result4 = detect_eye(nadia)
plt.imshow(result4,cmap='gray')
```

```python
result4 = detect_eye(denis)
plt.imshow(result4,cmap='gray')
```

```python
### Eye detection

cap = cv2.VideoCapture(0)

while True:
    
    ret,frame = cap.read(0)
    
    frame = detect_eye(frame)
    
    cv2.imshow('Video Face detect',frame)
    
    k = cv2.waitKey(1)
    
    if k == 27:
         break

cap.release()
cv2.destroyAllWindows()
```

```python
## Face detection
cap = cv2.VideoCapture(0)

while True:
    
    ret,frame = cap.read(0)
    
    frame = detect_img(frame)
    
    cv2.imshow('Video Face detect',frame)
    
    k = cv2.waitKey(1)
    
    if k == 27:
         break

cap.release()
cv2.destroyAllWindows()
```
